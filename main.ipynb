```python
# Tysiac

import random

symbols = {
    "Kier": "♥",
    "Karo": "♦",
    "Trefl": "♣",
    "Pik": "♠"
}

color_order = {
    "Kier": 0,
    "Pik": 1,
    "Karo": 2,
    "Trefl": 3
}

figure_order = {
    "A": 0,
    "10": 1,
    "K": 2,
    "D": 3,
    "J": 4,
    "9": 5
}

colors = ("Kier", "Karo", "Trefl", "Pik")
figures = ("9", "10", "J", "D", "K", "A")
cards = [(figure, color) for figure in figures for color in colors]

def print_cards(cards):
    for card in cards:
        print(card[0], end = " ")
        if card[1] == "Kier" or card[1] == "Karo":
            print("\033[31m", end="")
        else:            
            print("\033[90m", end="")
        print(symbols[card[1]], end=" ")
        print("\033[0m", end="")
    print()    

class player:
    def __init__(self, name):
        self.name = name
        self.cards = []
        self.score = 0
    


class game:
    round = 1
    players = [None, None, None, None]
    def __init__(self, player1):
        self.players[0] = player1
        self.players[1] = player("Player 2")
        self.players[2] = player("Player 3")
        self.players[3] = player("table")

    def deal_cards(self):
        random.shuffle(cards)
        self.players[0].cards = []
        self.players[1].cards = []
        self.players[2].cards = []
        self.players[3].cards = []
        i = 0 
        for j in range(3):
            self.players[3].cards.append(cards[i])
            i += 1
        for j in range(7):
            self.players[0].cards.append(cards[i])
            i += 1
        for j in range(7):
            self.players[1].cards.append(cards[i])
            i += 1
        for j in range(7):
            self.players[2].cards.append(cards[i])
            i += 1
        self.players[0].cards.sort(key=lambda card: (color_order[card[1]], figure_order[card[0]]))
        self.players[1].cards.sort(key = lambda card: (color_order[card[1]], figure_order[card[0]]))
        self.players[2].cards.sort(key = lambda card: (color_order[card[1]], figure_order[card[0]]))
        self.players[3].cards.sort(key = lambda card: (color_order[card[1]], figure_order[card[0]]))
        print_cards(self.players[0].cards)
        print_cards(self.players[1].cards)
        print_cards(self.players[2].cards)
        print_cards(self.players[3].cards)

    def bidding_phase(self):
        highest_bid = 100
        highest_bidder = self.players[self.round%3]
        bidder = (self.round%3)+1
        passed_players = []
        while len(passed_players) < 3:
            if bidder in passed_players:
                bidder = (bidder+1)%3
                continue
            bet = input("Current bet: " + str(highest_bid) + "\n" + self.players[bidder].name + ", enter your bet increase (or '0' to pass): ")
            if bet == "0":
                passed_players.append(bidder)
            else:
                highest_bid += int(bet)
                highest_bidder = self.players[bidder]
            bidder = (bidder+1)%3
        print("Highest bidder: " + highest_bidder.name + " with a bet of " + str(highest_bid))
        






        
def card_value(card):
    if card[0] == "9":
        return 0
    elif card[0] == "1":
        return 10
    elif card[0] == "J":
        return 2
    elif card[0] == "D":
        return 3
    elif card[0] == "K":
        return 4
    elif card[0] == "A":
        return 11
    else:
        return 0

def main():
    player1 = player("Mathias")
    game1 = game(player1)
    game1.deal_cards()
    game1.bidding_phase()

if __name__ == "__main__":
    main()
```
